```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from utils.Fingerprinter import Fingerprinter
```

```python
sns.set_context('paper',font_scale=1.5)
sns.set_style('whitegrid')
```

# Load data

```python
data = pd.read_csv('../data/FXA_CHEMBL.csv')
data.head()
```

```python
data['class_label'] = data['pChEMBL Value'].apply(lambda x : 1 if (x>=6.5) else 0)
```

```python
data
```

# Generate features

```python
feat_generator = Fingerprinter(smiles=data['Smiles'])
X = feat_generator.generate_fingerprint('ecfp',nbits=1024,radius=2)
```

```python
X
```

```python
X.shape
```

# Chemical space analysis

## PCA

```python
from sklearn.decomposition import PCA, KernelPCA
```

```python
pca = PCA(n_components=2)
kpca = KernelPCA(n_components=2,kernel='rbf')
```

```python
X_pca = pca.fit(X).transform(X)
X_kpca = kpca.fit(X).transform(X)
```

```python
pca.explained_variance_ratio_.sum()
```

```python
data.columns
```

```python
plt.figure(figsize=(12,9))
ax=sns.scatterplot(x=X_pca[:,0],y=X_pca[:, 1],s=120,edgecolor='k',hue=data['class_label'].values,palette='colorblind')


# Colorbar
#norm = plt.Normalize(data['class_label'].min(), data['class_label'].max())
#sm = plt.cm.ScalarMappable(cmap="magma", norm=norm)
#sm.set_array([])
#ax.figure.colorbar(sm)

ax.set_title('PCA (n_components = 2)',fontweight='bold')

sns.despine(right=True,top=True)

for _,s in ax.spines.items():
    s.set_linewidth(1.5)
    s.set_color('k')
```

```python
plt.figure(figsize=(12,9))
ax=sns.scatterplot(x=X_kpca[:,0],y=X_kpca[:, 1],s=120,edgecolor='k',hue=data['class_label'].values,palette='colorblind')


# Colorbar
#norm = plt.Normalize(data['class_label'].min(), data['class_label'].max())
#sm = plt.cm.ScalarMappable(cmap="magma", norm=norm)
#sm.set_array([])
#ax.figure.colorbar(sm)

ax.set_title('Kernel PCA (n_components = 2)',fontweight='bold')

sns.despine(right=True,top=True)

for _,s in ax.spines.items():
    s.set_linewidth(1.5)
    s.set_color('k')
```

# UMAP

```python
import umap
```

```python
reducer = umap.UMAP(n_neighbors=5, min_dist=0.8, metric='jaccard')
```

```python
embeddings = reducer.fit_transform(X)
```

```python
plt.figure(figsize=(12,9))
ax=sns.scatterplot(x=embeddings[:,0],y=embeddings[:, 1],s=120,edgecolor='k',hue=data['class_label'].values,palette='colorblind')


# Colorbar
# norm = plt.Normalize(data['pChEMBL Value'].min(), data['pChEMBL Value'].max())
# sm = plt.cm.ScalarMappable(cmap="magma", norm=norm)
# sm.set_array([])
# ax.figure.colorbar(sm)

ax.set_title('UMAP (n_neighbors = 20; min_dist=0.30)',fontweight='bold')

sns.despine(right=True,top=True)

for _,s in ax.spines.items():
    s.set_linewidth(1.5)
    s.set_color('k')
```

# t-SNE

```python
from sklearn.manifold import TSNE
```

```python
tsne_reducer = TSNE(init='pca',n_components=2,metric='jaccard')
```

```python
X_embedded = tsne_reducer.fit_transform(X)
```

```python
plt.figure(figsize=(12,9))
ax=sns.scatterplot(x=X_embedded[:,0],y=X_embedded[:, 1],s=120,edgecolor='k',hue=data['class_label'].values,palette='colorblind')


# Colorbar
# norm = plt.Normalize(data['pChEMBL Value'].min(), data['pChEMBL Value'].max())
# sm = plt.cm.ScalarMappable(cmap="magma", norm=norm)
# sm.set_array([])
# ax.figure.colorbar(sm)

ax.set_title('t-SNE (init = pca)',fontweight='bold')

sns.despine(right=True,top=True)

for _,s in ax.spines.items():
    s.set_linewidth(1.5)
    s.set_color('k')
```
